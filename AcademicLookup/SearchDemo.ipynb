```python
%%capture

from google.colab import drive
drive.mount('/content/drive')

from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

%cd /content/drive/MyDrive/AWCA/Colab_notebooks/CitationTagging/Sp21/CitationParser/AcademicLookup/

!pip install scholarly

import pandas as pd
from utils import get_df
import pandas as pd
import random

from online_resources import CompoundSearch
from utils import get_extensions, get_search_query

gc = gspread.authorize(GoogleCredentials.get_application_default())
```

```python
catalog_titles = [
    'catalog.20200407.' + suffix
    for suffix in ('aa', 'ab', 'ac', 'ad', 'ae')
]
df = pd.concat(
    (get_df(title, gc) for title in catalog_titles[:1]),
    ignore_index=True
)
df.head()
```

```python
extensions = get_extensions(df.Name)

for name in random.sample(list(df.Name), 10):
  print('Showing results for:', name)
  print(CompoundSearch(get_search_query(name, extensions), verbose=True).result())
  print()
```
