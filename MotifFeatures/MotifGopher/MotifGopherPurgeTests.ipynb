```python
import os
import random
import pickle
import matplotlib.pyplot as plt
import numpy as np

from MotifGopher import MotifGopher

N = 5000

with open(os.path.join('..', '..', 'datasets', 'occ_45K.pickle'), 'rb') \
        as dbfile:
    texts = random.sample(
        list(pickle.load(dbfile).raw),
        N
    )
```

```python
gopher3 = MotifGopher(
    texts,
    thresh=0.1,
    max_corr=0.9,
    confidence=0.99,
    r=0.75
)
```

```python
gopher3.plot_hunt()
```

```python
before_purge = dict(gopher3.found_motifs)
gopher3.purge(verbose=True)
```
