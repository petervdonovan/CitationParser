# MotifGopher Run
Here, I run the MotifGopher.

```python
import os
import random
import pickle
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from MotifGopher import MotifGopher

with open(os.path.join('..', '..', 'datasets', 'occ_45K.pickle'), 'rb') \
        as dbfile:
    texts = list(pickle.load(dbfile).raw)
```

Here I set the parameters that largely determine which and how many motifs will be found.

```python
gopher = MotifGopher(
    texts,
    thresh=0.1,
    max_corr=0.9,
    confidence=0.99,
    r=0.75
)
```

```python
gopher.run()
```
