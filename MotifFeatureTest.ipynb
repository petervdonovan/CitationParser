# MotifFeatureTest

Ad hoc testing for the MotifFeature class.

```python
import pickle
import os
from MotifFeatures.MotifFeature import MotifFeature
```

```python
db_path = os.path.join('datasets', 'occ_45K_train.pickle')
with open(db_path, 'rb') as dbfile:
    db = pickle.load(dbfile)
db.head(2)
```

```python
feature = MotifFeature(', ', 2)
for raw in db.sample(20).raw:
    out = feature.featurize(raw)
    if not all(val in (-float('inf'), float('inf')) for val in out):
        print(raw)
        print(list(out))
```
